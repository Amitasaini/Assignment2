Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data. Webpage https://www.naukri.com/

```python
# Importing necessary Libraries

import selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service

s=Service("C:/Users/Downloads/chromedriver.exe")
driver=webdriver.Chrome(service=s)
print(driver)
import pandas as pd
from selenium.webdriver.common.by import By 
import time
```

```python
# Open the url https://www.naukri.com/

url = "https://www.naukri.com/"
driver.get(url)
```

```python
# Write on Job title field 

job_tag = driver.find_element(By.CLASS_NAME,"suggestor-input")
job_tag.send_keys("Data Analyst")
```

```python
# Write on job location field

job_loc = driver.find_element(By.XPATH,"/html/body/div[1]/div[6]/div/div/div[5]/div/div/div/input")
job_loc.send_keys("Bangalore")
```

```python
# Click on search button

search_click = driver.find_element(By.CLASS_NAME,"qsbSubmit")
search_click.click()
```

```python
# Empty Lists

job_title = []
job_location = []
company_name = []
experience_required = []
```

```python
# Job title

title = driver.find_elements(By.XPATH,'//a[@class="title fw500 ellipsis"]')

for i in title[0:10]:
    job_title.append(i.text)
job_title
```

```python
# Job Location

location = driver.find_elements(By.XPATH,"//li[@class='fleft grey-text br2 placeHolderLi location']")

for i in location[0:10]:
    job_location.append(i.text)
job_location
```

```python
# Company Name

company = driver.find_elements(By.XPATH,"//a[@class='subTitle ellipsis fleft']")

for i in company[0:10]:
    company_name.append(i.text)
company_name
```

```python
# Experience Required

experience = driver.find_elements(By.XPATH,"//li[@class='fleft grey-text br2 placeHolderLi experience']//span")

for i in experience[0:10]:
    experience_required.append(i.text)
experience_required
```

```python
First_10_Jobs_Data_Analyst = pd.DataFrame(index = list(range(1,11)))

First_10_Jobs_Data_Analyst["Job Title"] = job_title[:10]
First_10_Jobs_Data_Analyst["Company Name"] = company_name[:10]
First_10_Jobs_Data_Analyst["Experience Required"] = experience_required[:10]
First_10_Jobs_Data_Analyst["Job Location"] = job_location[:10]

First_10_Jobs_Data_Analyst
```

Q.4 Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:
1. Brand
2. Product Description
3. Price

```python
# Importing necessary Libraries

import selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service

s=Service("C:/Users/Downloads/chromedriver.exe")
driver=webdriver.Chrome(service=s)
print(driver)
import pandas as pd
from selenium.webdriver.common.by import By 
import time
```

```python
# Let's open the url https://www.flipkart.com/

url = "https://www.flipkart.com/"
driver.get(url)
```

```python
# Search the product name

title = driver.find_element(By.CLASS_NAME,"_3704LK")
title.send_keys("sunglasses")
```

```python
# Click on search box

search_click = driver.find_element(By.CLASS_NAME,"L0Z3Pu")
search_click.click()
```

```python
# Empty Lists

url_pages = []
brand_name = []
product_description = []
price = []
discount = []
```

```python
# Page Url's (to Click on page numbers one by one)

pages = driver.find_elements(By.XPATH,'//nav[@class="yFHi8N"]//a') # a is becasue all url's is in a class
for i in pages:
    url_pages.append(i.get_attribute("href"))
url_pages
```

```python
# Brand names

for i in url_pages[:3]:     # [:3] means I selected 3 pages because I want to fetch only 100 details and 1 page has 40 details
    driver.get(i)
    brand = driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    for j in brand:
        brand_name.append(j.text)
    
brand_name
```

```python
# Product Description

for i in url_pages[:3]:    
    driver.get(i)
    product = driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
    for j in product:
        product_description.append(j.text)
    
product_description
```

```python
# Price

for i in url_pages[:3]:    
    driver.get(i)
    pri = driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
    for i in pri:
        price.append(i.text)
    
price
```

```python
# Discount

for i in url_pages[:3]:    
    driver.get(i)
    dis = driver.find_elements(By.XPATH,'//div[@class="_3Ay6Sb"]')
    for i in dis:
        discount.append(i.text)
    
discount
```

```python
Flipcart_100_Sunglasses = pd.DataFrame(index = list(range(1,101)))

Flipcart_100_Sunglasses["Brand"] = brand_name[:100]
Flipcart_100_Sunglasses["Product Description"] = product_description[:100]
Flipcart_100_Sunglasses["Price"] = price[:100]
Flipcart_100_Sunglasses["Discount"] = discount[:100]

Flipcart_100_Sunglasses
```

Q.5 Scrape 100 reviews data from flipkart.com for iphone11 phone. You have to go the link:
https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART 
As shown in page you have to scrape the tickmarked attributes. These are: 1 Rating 2 Review summary 3 Full review 4 You have to scrape this data for first 100 reviews.

```python
# Importing necessary Libraries

import selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service

s=Service("C:/Users/Downloads/chromedriver.exe")
driver=webdriver.Chrome(service=s)
print(driver)
import pandas as pd
from selenium.webdriver.common.by import By 
import time
```

```python
# Let's open the given url...

url = 'https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART'
driver.get(url)
```

```python
# Empty lists

url_page = []
rating = []
review_summary = []
full_review = []
```

```python
# Click on All Reviews 

pages = driver.find_element(By.CLASS_NAME,"gUuXy-")
pages.click()
```

```python
# All pages links

urls = driver.find_elements(By.XPATH,'//nav[@class="yFHi8N"]//a')
for i in urls:
    url_page.append(i.get_attribute("href"))
url_page
```

```python
# Ratings

for i in url_page:
    driver.get(i)
    ratings = driver.find_elements(By.XPATH,'//div[@class="_3LWZlK _1BLPMq"]')
    for i in ratings:
        rating.append(i.text)
rating
```

```python
# Review Summary

for i in url_page:
    driver.get(i)
    summary = driver.find_elements(By.XPATH,'//p[@class="_2-N8zT"]')
    for i in summary:
        review_summary.append(i.text)
review_summary
```

```python
# Full Review

for i in url_page:
    driver.get(i)
    full = driver.find_elements(By.XPATH,'//div[@class="t-ZTKy"]')
    for i in full:
        full_review.append(i.text.replace("\n",""))
full_review
```

```python
# Data Frame

Data_100_reviews = pd.DataFrame(index = list(range(1,101)))

Data_100_reviews["Ratings"] = rating[:100]
Data_100_reviews["Review Summary"] = review_summary[:100]
Data_100_reviews["Full review"] = full_review[:100]


Data_100_reviews
```
